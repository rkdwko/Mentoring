```python
import pandas as pd
import numpy as np 

l = pd.DataFrame({'key':['k0','k4','k2','k3'],
                     'A':['a0','a1','a2','a3'],
                     'B':['b0','b1','b2','b3']})
r = pd.DataFrame({'key':['k0','k1','k2','k3'],
                     'C':['c0','c1','c2','c3'],
                     'D':['d0','d1','d2','d3']})
```

```python
l
```

```python
r
```

```python
pd.merge(l,r,on='key')
```

```python
pd.merge(l,r,how='left',on='key')
```

```python
pd.merge(l,r,how='right',on='key')
```

```python
pd.merge(l,r,how='outer',on='key')
```
