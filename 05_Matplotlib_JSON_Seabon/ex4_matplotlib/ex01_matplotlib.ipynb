```python
import numpy as np 
t = np.arange(0,12,0.01)  
y = np.sin(t)

plt.figure(figsize=(10,6)) 
plt.plot(t,y) 
plt.show()
```

```python
plt.figure(figsize=(10,6)) 
plt.plot(t,y) 
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6)) 
plt.plot(t,y) 
plt.xlabel("time")
plt.ylabel("Amp")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6)) 
plt.plot(t,y) 
plt.xlabel("time")
plt.ylabel("Amp")
plt.title("Example of sinewave ")
plt.grid()
plt.show()
```
