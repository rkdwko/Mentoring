## merge 상권영역 + 유동인구

```python
population = pd.read_table("C:/Users/709-000/Desktop/새 폴더/연도별 추출(2018)/유동인구2018test2.txt",sep=',')
population
```

```python
del population['Unnamed: 0']
```

```python
population.head(10)
```

```python
population.columns
```

```python
population_1 = population
population_1.columns
```

```python
population_2 = population_1.groupby('trdar_cd').mean()
```

```python
population_2.columns
```

```python
del population_2['stdr_qu_cd']
population_2.head(10)
```

```python
del population_2['stdr_yy_cd']
population_2.head(10)
```

```python
population_2 = population_2.reset_index()
population_2.head(10)
```

```python
population_2 = population_2.rename(columns = {'trdar_cd':'상권_코드'}) # key값 통일
population_2.head(10)
```

```python
p_a_merge = pd.merge(population_2,area, on = '상권_코드')
p_a_merge.head(10)
```
